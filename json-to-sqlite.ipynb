```python
import json
import sqlite3
```

```python
conn = sqlite3.connect("data.db")
cur = conn.cursor()
```

```python
cur.execute("""
    CREATE TABLE IF NOT EXISTS files (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        title TEXT NOT NULL,
        type TEXT NOT NULL,
        season INTEGER,
        episode INTEGER,
        quality TEXT,
        language TEXT,
        format TEXT NOT NULL,
        url TEXT NOT NULL
    );
""")
cur.execute("CREATE INDEX IF NOT EXISTS idx_files_title ON files(LOWER(title));")
```

```python
with open("final.json", "r") as file:
    data = json.loads(file.read())
    print("Saving data to sqlite database...")
    for som, names in data.items():
        for name, sobs in names.items():
            for sob, qualities in sobs.items():
                for quality, seasons in qualities.items():
                    for season, episodes in seasons.items():
                        for episode, urls in episodes.items():
                            for url in urls:
                                file_format = url.split(".")[-1]
                                typee = ""
                                if file_format in ["jpg", "jpeg", "png"]:
                                    typee = "picture"
                                if file_format in ["mkv", "mp4"]:
                                    typee = "video"
                                elif file_format in ["mka", "mp3"]:
                                    typee = "audio"
                                elif file_format in ["zip", "rar"]:
                                    typee = "archive"
                                elif file_format in ["srt", "ssa"]:
                                    typee = "subtitle"

                                cur.execute("""
                                    INSERT INTO files (title, type, season, episode, quality, language, format, url)
                                    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
                                """, (name, som, season, episode, quality, sob, typee, url))
conn.commit()
print("Done.")
```
